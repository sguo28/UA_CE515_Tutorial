### Numpy
Numpy is a powerful package for scientific computing. We will mainly use numpy to generate random samples in this lecture. 
Tutorials of numpy can be found here: https://numpy.org/doc/stable/

Usage: **'import numpy as np'**

Basic data structures: numpy array, numpy ndarray.  
Basic data types: numerical types such as int, float, bool

Check this link for common operation routines using numpy:
https://numpy.org/doc/stable/reference/routines.html

```python
import numpy as np
#create 1-d array and conversion to and from list
mylist=[1,2,3,4,5]
myarray=np.array(mylist)

print(myarray)
print(f"length of array: {myarray.shape}")
print(myarray.dtype.type)
```

```python
#numpy cannot contain mixed data types
mylist=[1,2,3,'a']
myarray=np.array(mylist)

print(myarray)
print(f"length of array: {myarray.shape}")
print(myarray.dtype.type)
```

```python
#N-D array, e.g., matrix
mylist=[[1,2,3,4,5]]

myarray=np.array(mylist)

print(myarray)
print(f"length of array: {myarray.shape}, axis1 length: {myarray.shape[0]}, axis2 length: {myarray.shape[1]}")
print(myarray.dtype.type)
```

```python
#transpose
myarray=myarray.T

print(myarray)
print(f"length of array: {myarray.shape}, axis1 length: {myarray.shape[0]}, axis2 length: {myarray.shape[1]}")
print(myarray.dtype.type)
```

```python
#1-d array initialization
myarray=np.zeros(5); print(myarray, myarray.shape)
myarray=np.ones(5);print(myarray, myarray.shape)


#Initialize nd-array   

#dimension of 5 by 3
print("\n=====================\ndimension of 5 by 3")
myarray=np.ones((5,3)); print(myarray, myarray.shape)

#identity matrix
print("\n=====================\nidentity matrix")
myarray=np.eye(5); print(myarray)

#random matrix (Value between 0,1)
print("\n=====================\nrandom matrix")
myarray=np.random.random((5,3)); print(myarray)

#random integer 
#input: low, high (not include), size

print("\n=====================\nrandom integer")
myarray=np.random.randint(0,3,size=(5,3)); print(myarray)
```

```python
#matrix operations 
a1=np.random.randint(0,3, size=(2,3))
a2=np.random.randint(0,3, size=(3,2))
a3=np.random.randint(0,3, size=(2,3))
print('a1\n',a1)
print('a3\n',a3)
print('Multipcation results of a1*a3\n',a1*a3)  ##<--- this is the dot product, two matrix must be of the same shape
#this is wrong 
#print(a1*a2) 

#matrix multiplycation
print('Results of matrix product of a1 and a2\n',np.matmul(a1,a2))  #results is 5 by 5
```

```python
#random number generation.... 
print('Generating a random number between 0 and 1: ', np.random.random())

print('Generating a random integer between 0 and 4: ', np.random.randint(0,5))

#Sampling.... 
print('Choose one element in a list at random:', np.random.choice([1.1,2.2,3.3]))
print('Choose multiple element in a list at random:', np.random.choice([1.1,2.2,3.3],size=5))

#choose with given probability 
print(np.random.choice([1.1,2.2,3.3],size=5,p=[0.3,0.4,0.3]))

#get the largest value from a np array
mydata=np.random.random((5))
print(f"mydata: {mydata}")
print(np.argmax(mydata)) #return the index of the largest element in mydata
```

### Geopandas

Geopandas is the go to package when you want to work on shapefile.
Shapefile contains spatial objects such as point, polyline, polygon. Geopandas is built on pandas and can be used to process, visualize shapefiles and perform spatial algorithms. Geopandas parse a shapefile into a DataFrame.

For tutorials, check: https://geopandas.org/

Usage:**'import geopandas as gpd'**

An example here showing how to use geopandas to work with road shapefile in Tuscaloosa County. 

```python
# first install py7zr
# conda install -c conda-forge py7zr

# use py7zr to extract the shapefile
import py7zr
with py7zr.SevenZipFile('data/Tuscaloosa_road.7z', mode='r') as z:
    z.extractall(path='data/Tuscaloosa_road/')
```

```python
import zipfile
with zipfile.ZipFile('data/cb_2018_us_zcta510_500k.zip', mode='r') as z:
    z.extractall(path='data/CBG/')
```

```python
import geopandas as gpd # process the shapefile
```

```python
df=gpd.read_file("data/Tuscaloosa_road/Tuscaloosa_County_Roads.shp") #read the shapefile
```

```python
df.head()
```

```python
#visualize
df.plot()
```

```python
#extract roads with certian length
df2=df[df['Shape__Len']>5000]
df2.plot()
```

```python
#color plot based on the length of road segments
df2.plot(column='Shape__Len',cmap='OrRd')
```

```python
# randomly assign integer values between 0 to 10 to each road segment 

df2['random']=np.random.randint(0,10,size=df2.shape[0])
```

### Visual the random values on the route network

- Source of color map : https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
# set up a blank map
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, figsize=(6, 4.5))

#visualize the random values
df2.plot(column='random',cmap='Oranges', legend=True, legend_kwds={'label': "Random Value"}, ax=ax) # set the range of colormap: vmin=0, vmax=10

plt.tight_layout()

#save the figure
plt.savefig('results/road_wl_random_color.png',dpi=300)
```
